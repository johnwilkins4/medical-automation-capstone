# Agentic Healthcare Assistant — Demo

End-to-end walkthrough of the capstone's flagship scenario:

> *"My 70-year-old father has chronic kidney disease. I want to book a nephrologist for him. Also, can you summarize latest treatment methods?"*

Before running, make sure you have:
1. Built the database: `python -m db.seed`
2. Built the FAISS indices: `python -m src.embeddings --build`
3. Set `OPENAI_API_KEY` in `.env`

```python
import os, sys
sys.path.insert(0, os.path.abspath('..'))
from src.agent import run_agent
```

## 1. The flagship scenario

```python
query = (
    "My 70-year-old father has chronic kidney disease. "
    "I want to book a nephrologist for him. "
    "Also, can you summarize latest treatment methods?"
)
state = run_agent(query)
print(state['final_answer'])
```

## 2. Inspect the plan

```python
import json
print(json.dumps(state['plan'], indent=2, default=str))
```

## 3. Step-by-step trace

```python
for s in state['step_results']:
    print(f"step {s['step']}: {s['tool']}  ({s['latency_ms']} ms, success={s['success']})")
    print(f"  args: {s['args']}")
    preview = json.dumps(s['output'], default=str)[:200]
    print(f"  output preview: {preview}\n")
```
